```python
회원카드 = df['회원카드번호'].unique()
행병합 = []

for i in 회원카드:
    a = df.query('회원카드번호 == @i').sort_values(by="연월")  # 연월에 대해 데이터프레임을 정렬
    a.reset_index(drop=True, inplace=True)

    for j in range(len(a)):
        if j == len(a)-1:
            continue
        if (a.loc[j, '등록명'] in 등록명) and (a.loc[j+1, '등록명'] not in 등록명):
            a.loc[j+1, '연월'] = a.loc[j, '연월']
            a.loc[j+1, '등록명'] = a.loc[j, '등록명']
        
    행병합.append(a)  # 행병합 리스트에 데이터프레임 추가

# 모든 데이터프레임을 행 방향으로 병합
df1 = concat(행병합, ignore_index=True)
df1
            
```

```python
from collections import Counter
import pandas as pd

def mode_func(x):
    if len(x) > 0:
        # 문자열 요일을 숫자로 매핑
        day_mapping = {'월': 1, '화': 2, '수': 3, '목': 4, '금': 5, '토': 6, '일': 7}
        numeric_days = x.map(day_mapping)
        
        # 숫자로 매핑된 요일 중 가장 많이 등장하는 요일 찾기
        most_common_day_numeric = Counter(numeric_days).most_common(1)[0][0]
        
        # 매핑된 요일을 다시 문자열로 변환
        day_mapping_reverse = {v: k for k, v in day_mapping.items()}
        if not pd.isnull(most_common_day_numeric):
            most_common_day = day_mapping_reverse[most_common_day_numeric]
            return most_common_day
        else:
            return None  # 또는 다른 기본값으로 대체할 수도 있습니다.


df14 = df13.groupby(['회원카드번호', '연월', '등록명']).agg({
    '환불내역': 'sum',
    '요일': mode_func,  
    '미수액': 'sum',
    '거래일_일': 'max',
    '환불일_일': 'max',
    '일일입장': 'sum',
    '분실': 'sum',
    '부가서비스사용': 'sum',
    '레슨여부': 'sum',
    '매출액': 'sum',
    '개월수': 'sum',
    '골프전환': 'sum',
    '단체회원': 'max',
    '골프매출': 'sum',
    '라카매출': 'sum',
    '헬스매출': 'sum',
    'PT매출': 'sum',
    '골프레슨매출': 'sum',
    '일일입장매출': 'sum',
    '양도비매출': 'sum',
    '필라테스매출': 'sum',
    '환불사용금매출': 'sum',
    '의류매출': 'sum',
    '분실매출': 'sum',
    'GX매출': 'sum',
    '상환액매출': 'sum',
    '입회비매출': 'sum',
    '예치금매출': 'sum'
})
df14.reset_index(inplace=True)
df14
```
