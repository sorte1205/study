```python
import re
import json
import time
import pandas as pd
import datetime as dt
import chromedriver_autoinstaller
import numpy as np
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from tqdm.notebook import tqdm
```

```python
# 실사용 할 무신사 URL
musinsaUrl = "https://www.musinsa.com/ranking/best?period=month&age=ALL&mainCategory=&subCategory=&leafCategory=&price=&golf=false&kids=false&newProduct=false&exclusive=false&discount=false&soldOut=false&page={}&viewType=small&priceMin=&priceMax="
pageNumber = range(1,101)
num_loops = 10
```

```python
# 한글 제거 함수
def remove_korean(text):
    pattern = re.compile('[가-힣]+')  # 한글 패턴
    result = pattern.sub('', text)
    return result
```

```python
musinsaData = []

chromedriver_autoinstaller.install()
driver = webdriver.Chrome()
driver.implicitly_wait(2)
for loop in range(num_loops):
  for p in pageNumber:
      url = musinsaUrl.format(p)
      # print(url)
      
      driver.get(url)
      driver.implicitly_wait(2)

      soup1 = BeautifulSoup(driver.page_source)

      itemList = soup1.select(".li_box")

      print_item_details = True 

 
#goodsRankList > li:nth-child(1) > div.li_inner > div.article_info > p.list_info > a
#goodsRankList > li:nth-child(2) > div.li_inner > div.article_info > p.list_info > a
#goodsRankList > li:nth-child(90) > div.li_inner > div.article_info > p.list_info > a
      i = 1
      while i <91:
        for li in itemList:
            itemNo = li['data-goods-no']
            # print(itemNo)

            
            title_f = "#goodsRankList > li:nth-child({}) > div.li_inner > div.article_info > p.list_info > a"
            reserve_f = "#goodsRankList > li:nth-child({}) > div.li_inner > div.article_info > p.list_info .txt_reserve"
            delay_f = "#goodsRankList > li:nth-child({}) > div.li_inner > div.article_info > p.list_info .txt_delay"
            
            title = title_f.format(i)
            reserve = reserve_f.format(i)
            delay = delay_f.format(i)


            title = soup1.select(title)
            reserve = soup1.select(reserve)
            delay = soup1.select(delay)      
            
            if title:
              titleValue = title[0].text.strip()
              titleValue = titleValue.replace('[무료반품]', '').strip()
            else:
              titleValue = ""
              # 상품명 가져오기

          
            #배송지연
            if reserve:
              reserveValue = reserve[0].text.strip()
            else:
              reserveValue = ""
              # print(reserveValue)
          
            #상품지연
            if delay:
              delayValue = delay[0].text.strip()
            else:
              delayValue = ""
              # print(delayValue)
          
            while reserveValue or delayValue in titleValue:
              if reserveValue not in titleValue:
                  pass
              elif delayValue not in titleValue:
                  pass
              else:
                  titleValue = titleValue.replace(reserveValue,'').strip()
                  titleValue = titleValue.replace(delayValue,'').strip()
                  break          
            i +=1

            mydict = {
              "상품번호": itemNo,
              "상품명": titleValue}
            
            musinsaData.append(mydict)

musinsaData
print("수집된 상품 갯수: ",len(musinsaData))

driver.quit()
```
