## #02. 선호연령비율 분할 

```python
# 연령 별 컬럼 생성
df = origin.filter(['상품명','선호연령비율'])
df['~18세'] =""
df['19~23세']=""
df['24~28세']=""
df['29~33세']=""
df['34~39세']=""
df['40세~']=""

# 선호연령비율 컬럼 내용을 리스트에서 문자열로 변환
raw=df['선호연령비율']
for i in range(len(df.index)):
    raw[i]=raw[i].strip("[]")

for j in range(len(df.index)):
    try:
        input_str = raw[j]

# 정규표현식을 사용하여 숫자와 연령대를 추출
        matches = re.findall(r'(\d+)?~?(\d+)세~?(\d+)?,', input_str)

# 추출된 정보를 리스트에 저장
        result = [int(match[2]) if match[2] else (int(match[0]) if match[0] else int(match[1])) for match in matches]
        raw[j] = result
    except Exception as e:
        print("예외발생", e)
error = []
for i in range(len(df.index)):
    if len(raw[i]) <6:
        error.append(i)
print("error index:", error, "총 error:", len(error))

for j in range(len(df.index)):
    if j in error:
        pass
    else:
        df['~18세'][j] =raw[j][0]
        df['19~23세'][j]=raw[j][1]
        df['24~28세'][j]=raw[j][2]
        df['29~33세'][j]=raw[j][3]
        df['34~39세'][j]=raw[j][4]
        df['40세~'][j]=raw[j][5]
```

```python
merged = merge(origin, df, on = '상품명')
merged
```

```python
merged.to_excel("musinsa_monthrank_data.xlsx", index=False)
```

## #03. 수치형 데이터 정제

```python
numeric_df = origin.filter(['상품명','조회수','누적판매','상품가격','~18세','19~23세','24~28세','29~33세','34~39세','40세~'])
numeric_df
```

### [1] 수치조정 (조회수, 누적판매, 상품가격,연령수치)

```python
numeric_df['조회수'] = numeric_df['조회수'] * 0.001
numeric_df['누적판매'] = numeric_df['누적판매'] * 0.001
numeric_df['상품가격'] = numeric_df['상품가격'] * 0.001

numeric_df
```

```python
numeric_df['~18세'] = numeric_df['~18세'] * 0.01
numeric_df['19~23세'] = numeric_df['19~23세'] * 0.01
numeric_df['24~28세'] = numeric_df['24~28세'] * 0.01
numeric_df['29~33세'] = numeric_df['29~33세'] * 0.01
numeric_df['34~39세'] = numeric_df['34~39세'] * 0.01
numeric_df['40세~'] = numeric_df['40세~'] * 0.01

numeric_df
```

```python
merged = merge(origin, numeric_df, on = '상품명')
merged.to_excel("musinsa_monthrank_data.xlsx", index=False)
```
