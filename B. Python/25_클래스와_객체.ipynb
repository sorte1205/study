## 1. 객체지향 프로그래밍의 이해
### (1) 객체지향 프로그래밍의 이해
- 객체지향 언어
  - 객체 중심 프로그래밍 패러다임
- 객체 (object)
  - 필요에 따라 다른 변수와 함수를 내장하고 있는 특수한 형태의 변수

    - 메서드: 객체 안에 포함된 함수
- 클래스 (class)
  - 객체의 설계도 역할을 하는 프로그램소스
  - 하나의 설계도를 통해 동일한 구조를 갖는 객체를 여러 개 생성할 수 있다.
- 객체지향 프로그래밍의 예시
  - 게임
    - 명사적 특성: hp, speed, dps
    - 동사적 특성: move, shield....

## 3. 멤버변수의 범위
### (1) 클래스 내의 변수 구분
- 지역변수와 전역변수
  - 클래스에 멤버변수와 메서드가 공존할 경우, 멤버변수는 모든 메서드가 공유하는 전역 변수로 존재하게 된다.
  - 변수의 종류
    - 멤버변수 (=전역변수)
      - 클래스 안에서 선언된 변수.
      - 클래스 블록 범위 안에서 유효
    - 지역변수
      - 메서드 안에서 선언된 변수
      - 메서드 블록을 빠져나가지 못함
      - 파라미터도 지역변수의 일종

```python
# 변수와 함수를 모두 내장하는 클래스 정의
class Member:
    username = None
    email = None

    def join(self, username, email):
        self.username = username
        self.email = email
    
    def view_info(self):
        print(self.username)
        print(self.email)

# 클래스의 기능을 객체에 부여하기
mem1 = Member()
mem1.join("leekh", "leekh@gmail.com")
mem1.view_info()
```
