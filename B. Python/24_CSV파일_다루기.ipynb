# 24 CSV 파일 다루기
## 1. CSV 파일의 이해
### (1) CSV 파일 살펴보기
### (2) CSV 파일 생성하기
- 엑셀에서 사용하는 표 형식을 텍스트 형태로 변환한 파일
  - 데이터 구조를 표현하는 가장 간결한 형태의 파일로서, 엑셀의 행은 한 줄로 표현하고 엑셀의 열은 콤마(,)로 구분하는 형식이다 (txt로 열면 알 수 있음)



## 2. CSV 파일 저장하기
### (1) 딕셔너리를 포함하는 리스트 구조를 CSV로 저장하기
- 초기화
  - csv 파일의 한 행에 대한 구문 형식 정의
- csv 파일 저장을 위한 f객체 생성
  - Excel의 csv는 euc-kr 형식
- 첫 줄에 각 항목의 제목 기록
- 각 데이터를 한 줄씩 콤마로 구분하여 기록
### (2) 다른 형식의 데이터를 CSV로 저장하기
  - 한 행의 템플릿 (초기화)
  - 딕셔너리의 키를 리스트로 변환
  - keys 리스트 사이에 쉼표를 넣어 문자열로 변환
  - 첫 줄에 각 행목 제목 기록
  - 각 데이터를 한 줄씩 콤마로 구분하여 기록




```python
# (1) 딕셔너리를 포함하는 리스트 구조를 CSV로 저장하기
grade = [
    {"name": "철수", 'kor':95, 'eng':88, 'math':72},
    {"name": "영희", 'kor':92, 'eng':98, 'math':95},
    {"name": "철민", 'kor':88, 'eng':76, 'math':64}
]

tpl= "{0},{1},{2},{3}\n"

with open("grade.csv","w", encoding = 'euc-kr') as f:
    f.write("이름, 국어, 영어, 수학\n")

    for item in grade:
        tmp = tpl.format(item["name"], item["kor"], item["eng"], item["math"])
        f.write(tmp)

```

```python
# (2) 다른 형식의 데이터를 CSV로 저장하기
grade_dic= {
    '이름': ['철수', '영희', '민철', '수현', '호영'],
    '국어': [98, 88, 92, 63, 120],
    '영어': [76, 90, 70, 60, 50],
    '수학': [88, 62, 88, 31, 76],
    '과학': [64, 72, 45, 70, 88]
}

tpl = "{0},{1},{2},{3},{4}\n"
keys = list(grade_dic.keys())
p = ","
title = p.join(keys)

with open("grade2.csv", "w", encoding = 'euc-kr') as f:
    f.write(title+"\n")

    for i in range(0, len(grade_dic['이름'])):
        tmp = tpl.format(grade_dic['이름'][i], grade_dic['국어'][i], grade_dic['영어'][i],
                         grade_dic['수학'][i], grade_dic['과학'][i])
        f.write(tmp)
```

## 3. CSV 파일 읽어오기

### (1) CSV파일의 내용을 파이썬 데이터로 변환 (Step-1)
- 읽어들인 내용을 저장하기 위한 빈 변수 생성
- grade.csv 파일에 대한 파일 객체를 읽기 모드로 생성
  - 파일의 각 행을 원소로 갖는 리스트 생성
  
### (2) CSV파일의 내용을 파이썬 데이터로 변환 (Step-2)
- CSV의 데이터를 저장할 빈 배열
- 첫 행은 제목이므로 건너뜀
- 현재 행의 내용을 콤마를 기준으로 잘라 리스트로 변환
  - split()의 결과로 생성된 리스트의 모든 원소는 string타입 (""가 삽입됨)
- 한 행의 내용을 딕셔너리로 변환
- 딕셔너리를 미리 준비한 리스트에 추가


```python
# (1) CSV파일의 내용을 파이썬 데이터로 변환 (Step-1)

csv_list = None

with open("grade.csv", "r", encoding='euc-kr') as f:
    csv_list = f.readlines()

print(csv_list)

# (2) CSV파일의 내용을 파이썬 데이터로 변환 (Step-2)
data = []
for i, line in enumerate(csv_list):
    if i == 0:
        continue # 제목은 저장 안할 것
    
    k = line.strip().split(",")
    
    item = {"name":k[0], "kor": int(k[1]), "eng": int(k[2]), "math":(int(k[3]))}

    data.append(item)

data
```
