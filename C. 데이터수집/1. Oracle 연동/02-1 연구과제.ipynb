# 연구과제

    https://data.hossam.kr/py/traffic_acc.csv 파일은 2005년도부터 2018년도까지의 교통사고 현황을 조사한 자료 입니다.
    이 파일을 자신의 작업폴더에 다운로드 받은 후 데이터베이스에 저장하세요.
    이를 위해 직접 DBMS안에 테이블을 생성해야 합니다.

### [1] 준비과정
(1) 테이블 정의         (sql developer에서 테이블 생성)
CREATE TABLE traffic_acc1 (
    id number not null,
    year number not null,
    month number not null,
    cases number not null,
    death number not null,
    injury number not null,
    primary key(id)
);

```python
# CSV 데이터 가져오기
traffic_acc = None
with open("traffic_acc.csv", "r", encoding='utf-8') as f:
    traffic_acc = f.readlines()

print(traffic_acc)
```

```python
# 데이터 베이스 접속
import cx_Oracle as cx
dbcon = cx.connect("hr", "hr", "localhost:1521/xe")
cursor = dbcon.cursor()
```

```python
# CSV에서 가져온 데이터 SQL로 저장하기
sql = "INSERT INTO traffic_acc1 (id, year, month, cases, death, injury) values (seq_traffic_acc.nextval, :1, :2, :3, :4, :5)"

for item in traffic_acc[1:]:
    line = item.strip().split(",")
    line = list(map(int,line))
    cursor.execute(sql,line)    
        

dbcon.commit()
```

```python
# 결과확인
sql = "SELECT * FROM traffic_acc1"
cursor.execute(sql)
cursor.rowfactory = lambda *args: dict(zip([d[0] for d in cursor.description], args))
result = cursor.fetchall()
result
```
