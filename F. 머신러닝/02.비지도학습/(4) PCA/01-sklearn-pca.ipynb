```python
# 연결된 모듈이 업데이트 되면 즉시 자동 로드함
%load_ext autoreload
%autoreload 2

from hossam.util import *
from hossam.plot import *
from hossam.analysis import *

# 주성분 분석 패키지
from sklearn.decomposition import PCA
```

### [2] 데이터 가져오기

```python
origin = my_read_excel("https://data.hossam.kr/mldata/iris.xlsx", info=False)
```

## #02. 차원 축소 적용

### [1] 독립변수만 추출

차원 축소는 독립변수에 대해서만 적용되는 기법이다.

```python
x = origin.drop("Species", axis=1)
x
```

### [2] 데이터 스케일링

주성분 분석에서는 데이터 표준화(MinMax, Standard)가 필수 과정이다.

```python
df = my_standard_scaler(x)
df
```

### [3] 차원축소

sklearn의 PCA 분석은 기존의 변수들을 토대로 n_component 수 만큼의 새로운 변수를 생성한다.

일반적으로 머신러닝의 학습 데이터를 생성하는 용도

#### (1) 차원 축소 수행

```python
estimator = PCA(n_components=2)
pca = estimator.fit_transform(df)
pca
```

#### (2) 차원 축소 결과를 데이터 프레임으로 구성

차원축소의 컬럼 이름은 일반적으로 `PC1`, `PC2` ... `PCn` 형식으로 지정한다.

```python
pca_df = DataFrame(pca, columns=["PC1", "PC2"])
pca_df['target'] = origin['Species']
pca_df
```

#### (3) 생성된 변수들의 설명력 확인

```python
estimator.explained_variance_ratio_
```
