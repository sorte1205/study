```python
data = get_data('juice')
data
```

```python
s = ClassificationExperiment()
s.setup(data, target = 'Purchase', session_id = 123)
```

```python
s.models()
```

```python
xgboost = s.create_model('xgboost')
```

```python
best = s.compare_models()
```

```python
s.evaluate_model(best)
```

```python
pred_holdout = s.predict_model(best)
pred_holdout
```
